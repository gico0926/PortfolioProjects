```python
import pandas as pd

data = pd.read_json(r'D:\專案\Projects\Youtube channel analysis\url_list.json')

df = pd.DataFrame(data)

df
```

```python
df.fillna(0, inplace=True)
print(df)
print(df.dtypes)
```

```python
df_nonull = df.dropna()

print(df_nonull)
print(df_nonull.dtypes)
```

```python
df_nodouble = df_nonull.drop_duplicates()
df_nodouble
```

```python
# Assuming you have a DataFrame named df
column_name = df_nodouble.columns[0]
print(column_name)
```

```python
df_final = df_nodouble[df_nodouble['urls'].str.contains('list=UULFmsX3g', na=False)]
df_final
```
